```python
def input_by_model_new(data):
    return (
        data.groupby(['input_name', 'model_name'])
        .agg({'run_id': 'nunique', 'name': 'nunique'})
        .reset_index()
        .rename(columns={'name': 'unique_questions'})
        .sort_values('run_id', ascending=False)
    )
```

```python
input_by_model_new(data)
```

```python
def sheet_answers(
        data, 
        sheet_name: str, 
        model_name: str = None
    ):
    return (
        data = data[data.name == sheet_name]
        data[].groupby(['input_name', 'model_name'])
        .agg({'run_id': 'nunique', 'name': 'nunique'})
        .reset_index()
        .rename(columns={'name': 'unique_questions'})
        .sort_values('run_id', ascending=False)
    )
```
